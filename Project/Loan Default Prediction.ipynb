```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
#from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def train(models):
    for (name,model) in models:
        model.fit(X_train,y_train)
        predictions = model.predict(X_test)

        accuracy = accuracy_score(y_test,predictions)
        precision = precision_score(y_test,predictions)
        recall = recall_score(y_test,predictions)
        f1 = f1_score(y_test,predictions)
        mcc = matthews_corrcoef(y_test,predictions)
        conf_matrix= confusion_matrix(y_test,predictions)

        print(name)
        print("Accuracy =",accuracy)
        print("Precision =",precision)
        print("Recall =",recall)
        print("F1 score =",f1)
        print("MCC =",mcc)
        print("confusion matrix =",conf_matrix)
        print()
        scores.append((name,accuracy,precision,recall,f1,mcc))
```

```python
models = []
scores = []
models.append(("Decision Tree",DecisionTreeClassifier(max_depth=10, min_samples_leaf=5, criterion="gini", random_state=3)))
models.append(("Random Forest",RandomForestClassifier(n_estimators=65, oob_score=True, n_jobs=-1, random_state=3, max_features = None, min_samples_leaf =1)))
models.append(("Bagging",BaggingClassifier(n_estimators=100, random_state=3)))
models.append(("AdaBoost",AdaBoostClassifier(n_estimators=350, random_state=3)))
models.append(("Gradient Boosting", GradientBoostingClassifier(n_estimators=350, random_state=3)))
#models.append(("SVM",SVC()))
models.append(("Logistic Regression", LogisticRegression()))
models.append(("KNN", KNeighborsClassifier(n_neighbors=7)))
models.append(("Naive Bayes", GaussianNB()))
#models.append(("XGBoost Classifier",XGBClassifier()))
```

```python
train(models)
```
